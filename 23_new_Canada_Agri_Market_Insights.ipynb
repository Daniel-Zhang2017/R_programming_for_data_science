## Problem 1

#### Create tables

Establish a connection **conn** to the RSQLite database **FinalDB.sqlite**, and create the following four tables.

1.  **CROP_DATA**
2.  **FARM_PRICES**
3.  **DAILY_FX**
4.  **MONTHLY_FX**

The previous practice lab will help you accomplish this.


```r
# Write your query here
conn <- dbConnect(RSQLite::SQLite(),"FinalD.sqlite")
#CROP_DATA:
df1 <- dbExecute(conn, 
                    "CREATE TABLE CROP_DATA (
                                      CD_ID INTEGER NOT NULL,
                                      YEAR DATE NOT NULL,
                                      CROP_TYPE VARCHAR(20) NOT NULL,
                                      GEO VARCHAR(20) NOT NULL, 
                                      SEEDED_AREA INTEGER NOT NULL,
                                      HARVESTED_AREA INTEGER NOT NULL,
                                      PRODUCTION INTEGER NOT NULL,
                                      AVG_YIELD INTEGER NOT NULL,
                                      PRIMARY KEY (CD_ID)
                                      )"
                    )

    if (df1 == -1){
        cat ("An error has occurred.\n")
        msg <- odbcGetErrMsg(conn)
        print (msg)
    } else {
        cat ("Table CROP_DATA was created successfully.\n")
    }

# FARM_PRICES table
df2 <- dbExecute(conn, 
                    "CREATE TABLE FARM_PRICES (
                                      CD_ID INTEGER NOT NULL,
                                      DATE DATE NOT NULL,
                                      CROP_TYPE VARCHAR(20) NOT NULL,
                                      GEO VARCHAR(20) NOT NULL,
                                      PRICE_PRERMT DECIMAL(10,2) NOT NULL,
                                      PRIMARY KEY (CD_ID)
                                      )"
                    )

if (df2 == -1){
    cat ("An error has occurred.\n")
    msg <- odbcGetErrMsg(conn)
    print (msg)
} else {
    cat ("Table FARM_PRICES was created successfully.\n")
}

# DAILY_FX table
df3 <- dbExecute(conn, 
                    "CREATE TABLE DAILY_FX (
                                      DFX_ID INTEGER NOT NULL,
                                      DATE DATE NOT NULL,
                                      FXUSDCAD DECIMAL(10,4) NOT NULL,
                                      PRIMARY KEY (DFX_ID)
                                      )"
                    )

if (df3 == -1){
    cat ("An error has occurred.\n")
    msg <- odbcGetErrMsg(conn)
    print (msg)
} else {
    cat ("Table DAILY_FX was created successfully.\n")
}

# MONTHLY_FX table
df4 <- dbExecute(conn, 
                    "CREATE TABLE MONTHLY_FX (
                                      DFX_ID INTEGER NOT NULL,
                                      DATE DATE NOT NULL,
                                      FXUSDCAD DECIMAL(10,4) NOT NULL,
                                      PRIMARY KEY (DFX_ID)
                                      )"
                    )

if (df4 == -1){
    cat ("An error has occurred.\n")
    msg <- odbcGetErrMsg(conn)
    print (msg)
} else {
    cat ("Table MONTHLY_FX was created successfully.\n")
}

```

## Problem 5

#### How many hectares of Rye were harvested in Canada in 1968?


```r
# Write your query here
query <- "
    SELECT 
        SUM(HARVESTED_AREA) as total_harvested_hectares,
        SUM(PRODUCTION) as total_production,
        AVG(AVG_YIELD) as avg_yield
    FROM CROP_DATA 
    WHERE CROP_TYPE = 'Rye' 
        AND GEO = 'Canada'
        AND strftime('%Y', YEAR) = '1968'
"

result <- dbGetQuery(conn, query)
result
```

## Problem 10

#### Rank the crop types harvested in Saskatchewan in the year 2000 by their average yield. Which crop performed best?


```r
# Write your query here
query <- "
    SELECT 
        CROP_TYPE,
        AVG(AVG_YIELD) as avg_yield
    FROM CROP_DATA 
    WHERE GEO = 'Saskatchewan' 
        AND YEAR LIKE '2000%'
    GROUP BY CROP_TYPE
    ORDER BY avg_yield DESC
"

crop_ranking <- dbGetQuery(conn, query)
crop_ranking
```

## Problem 11

#### Rank the crops and geographies by their average yield (KG per hectare) since the year 2000. Which crop and province had the highest average yield since the year 2000?


```r
# Write your query here
# Query to rank crops and geographies by average yield since 2000
query <- "
    SELECT 
        CROP_TYPE,
        GEO,
        AVG(AVG_YIELD) as avg_yield
    FROM CROP_DATA 
    WHERE strftime('%Y', YEAR) >= '2000'
    GROUP BY CROP_TYPE, GEO
    ORDER BY avg_yield DESC
"

# Execute query
ranking_results <- dbGetQuery(conn, query)

# Display results
cat("🌾 Crop and Geography Rankings by Average Yield (Since 2000)\n")
cat("============================================================\n\n")
ranking_results
```

## Problem 12

#### Use a subquery to determine how much wheat was harvested in Canada in the most recent year of the data.


```r
# Write your query here
query <- "
    SELECT 
        GEO,
        SUM(HARVESTED_AREA) as harvested_area,
        SUM(PRODUCTION) as production,
        AVG(AVG_YIELD) as avg_yield

    FROM CROP_DATA 
    WHERE CROP_TYPE = 'Wheat' 
        AND YEAR = (
            SELECT MAX(YEAR) 
            FROM CROP_DATA 
            WHERE CROP_TYPE = 'Wheat'
        )
    AND GEO = 'Canada'
    GROUP BY GEO
    ORDER BY production DESC
"

provincial_breakdown <- dbGetQuery(conn, query)

# Display provincial breakdown
cat("\n📊 Provincial Breakdown of Wheat Production (Most Recent Year)\n")
cat("===============================================================\n\n")
provincial_breakdown
```

## Problem 13

#### Use an implicit inner join to calculate the monthly price per metric tonne of Canola grown in Saskatchewan in both Canadian and US dollars. Display the most recent 6 months of the data.


```r
# Write your query here
query <- "
    SELECT 
        f.DATE,
        f.CROP_TYPE,
        f.GEO,
        f.PRICE_PRERMT as price_cad,
        ROUND(f.PRICE_PRERMT / d.FXUSDCAD, 2) as price_usd,
        d.FXUSDCAD as exchange_rate
    FROM FARM_PRICES f, DAILY_FX d
    WHERE f.CROP_TYPE = 'Canola' 
        AND f.GEO = 'Saskatchewan'
        AND f.DATE = d.DATE
    ORDER BY f.DATE DESC
    LIMIT 6
"

# Execute query
recent_prices <- dbGetQuery(conn, query)

# Display results
cat("🌾 Canola Prices in Saskatchewan - Most Recent 6 Months\n")
cat("=========================================================\n\n")

print(recent_prices)
```
